## Customer churn model training for custoner retention

### Load Dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
churn_df = pd.read_csv('Telco-Customer-Churn.csv')
```

```python
churn_df.head()
```

```python
# Setting up X and y variables from dataset
X = pd.get_dummies(churn_df.drop(['customerID', 'Churn'], axis = 1))
y = churn_df.Churn.apply(lambda x : 1 if x=='Yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

### Import Dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

### Build Model

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['Accuracy'])
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
accuracy_score(y_test, y_hat)
```

The accuracy score seen here is 75% for it's predictions on customer churn. This is a reasonably accuracy given the difficulty of the dataset. While this model trains reasonably quickly, it could also potentially be improved further by increasing the epochs, or making a deeper layer neural network at the expense of training time.

### LinearSVC and LogisticRegression for comparison.

To illustrate the difference in the TF neural network here vs other models, LinearSVC and LogisticRegression have been imported below for comparison.

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
clf = make_pipeline(StandardScaler(), LinearSVC(random_state=0, tol=1e-5, max_iter=10000))
clf.fit(X_train, y_train)
```

```python
clf_y_hat = clf.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state=0, solver='lbfgs').fit(X_train, y_train)
log_reg.predict(X_test)
```

As can be seen, both of these model have difficultly converging due to the difficulty in predicting on this dataset. By comparison, TensorFlows Sequential model generates quite accurate predictions.
